```python
import os
import csv
import pandas as pd
from pandas_profiling import ProfileReport
from sklearn.decomposition import FactorAnalysis
```

```python
df_scores = pd.read_csv(os.path.join('data', '12_subject_scores.csv'), sep='\t')
```

```python
df_scores
```

```python
display(df_scores.describe())
```

```python
n_factors = 3
fa = FactorAnalysis(n_components=n_factors, random_state=57)
fa.fit(df_scores)
```

```python
df_factor_loading = pd.DataFrame(fa.components_.T, columns=[f'factor_{i}' for i in range(n_factors)], index=df_scores.columns)

df_factor_loading
```

factor analysis with varimax rotation

```python
from factor_analyzer import FactorAnalyzer
```

```python
fa_varimax = FactorAnalyzer(rotation='varimax',  n_factors=n_factors)
```

```python
fa_varimax.fit(df_scores)
```

```python
pd.DataFrame(fa_varimax.loadings_, index=df_scores.columns, columns=[f'factor_{i}' for i in range(n_factors)])
```

```python
fa_promax = FactorAnalyzer(rotation='promax', n_factors=n_factors)
```

```python
fa_promax.fit(df_scores)
```

```python
pd.DataFrame(fa_promax.loadings_, index=df_scores.columns, columns=[f'factor_{i}' for i in range(n_factors)])
```
